```python
# change standard charge column names to match
dcm.rename(columns={'standard_charge_negotiated_dollar': 'standard_charge_dollar'}, inplace=True)
dcm.rename(columns={'standard_charge_negotiated_percentage': 'standard_charge_percentage'}, inplace=True)
dcm.rename(columns={'standard_charge_negotiated_algorithm': 'standard_charge_algorithm'}, inplace=True)
dcm.rename(columns={'standard_charge_methodology': 'methodology'}, inplace=True)

hcm.rename(columns={'minimum': 'standard_charge_min'}, inplace=True)
hcm.rename(columns={'maximum': 'standard_charge_max'}, inplace=True)
```

```python
# compare where dcm[code_1_type] is the same as hcm[type] and dcm[code_1] is the same as hcm[code]
# only keep the standard charge columns as well as max min and hospital name
dcm_compare = dcm.merge(hcm, left_on=['code_1_type', 'code_1'], right_on=['type', 'code'], how='inner')


```

```python
# compare the prices between the two hospitals
def compare_prices(hcm, dcm):
    # Merge the two dataframes on the 'service' column
    merged = dcm.merge(hcm, left_on=['code_1_type', 'code_1'], right_on=['type', 'code'], how='inner', suffixes=('_dcm', '_hcm'))

    # Calculate the price differences
    merged['price_difference_max'] = merged['standard_charge_max_hcm'] - merged['standard_charge_max_dcm']
    merged['price_difference_min'] = merged['standard_charge_min_hcm'] - merged['standard_charge_min_dcm']
    merged['price_difference_estimated'] = merged['estimated_amount_hcm'] - merged['estimated_amount_dcm']

    return merged
```

```python
comparison_df = compare_prices(hcm, dcm)
```

```python
comparison_df
```

```python
# keep only the relevant columns    
comparison_df = comparison_df[['description_dcm','description_hcm', 'code_1_type', 'code_1', 'type', 'code', 'price_difference_max', 'price_difference_min', 'price_difference_estimated',]]
```

```python
comparison_df

# it's HCM - DCM meaning a positive value means HCM is more expensive than DCM and a negative value means DCM is more expensive than HCM
```

```python
import pprint

# For each price difference column, get the max, min, and mean, and the corresponding description_dcm

result = {}

for col in ['price_difference_max', 'price_difference_min', 'price_difference_estimated']:
    # Drop NaN for accurate idxmax/idxmin
    non_null = comparison_df.dropna(subset=[col])
    max_idx = non_null[col].idxmax()
    min_idx = non_null[col].idxmin()
    mean_val = non_null[col].mean()
    # Find the row closest to the mean
    mean_idx = (non_null[col] - mean_val).abs().idxmin()
    result[col] = {
        'max': {
            'value': non_null.loc[max_idx, col],
            'description_dcm': non_null.loc[max_idx, 'description_dcm']
        },
        'min': {
            'value': non_null.loc[min_idx, col],
            'description_dcm': non_null.loc[min_idx, 'description_dcm']
        },
        'mean': {
            'value': mean_val,
            'description_dcm': non_null.loc[mean_idx, 'description_dcm']
        }
    }

pprint.pprint(result)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(comparison_df['price_difference_estimated'].dropna(), bins=100, kde=True)
plt.title('Estimated Price Difference Between Hospitals by Procedure')
plt.xlabel('Estimated Price Difference (HCM - DCM)')
plt.ylabel('Number of Procedures')
plt.tight_layout()
plt.show()
```
